```python
import numpy as np
import csv
```

```python
n=50
mu,sigma=0,0.1
```

```python
header = ['x','y','f']
```

```python
with open('syntheticdatasetd2.csv','w',newline='') as file:
  writer=csv.writer(file)
  writer.writerow(header)
  for i in range(1,n):
    for j in range(1,n):
      noise = np. random. normal(mu,sigma,1)
      target=4*i*i*i-3*i*j+ noise
      [writer.writerow([i,j,target[0]])]
```

```python
import pandas as pd

df = pd.read_csv('syntheticdatasetd2.csv', parse_dates=False)
df.y.plot(color='g',lw=1.3)
df.f.plot(color='r',lw=1.3)
```

```python
import matplotlib.pyplot as plt
from chart_studio import plotly
from mpl_toolkits.mplot3d import axes3d
from IPython.display import HTML
from matplotlib import animation

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
with open('syntheticdatasetd2.csv', 'r') as f:
    obj = csv.reader(f)
    X, Y, Z = [], [], []
    for i,row in enumerate(obj):
        if i>0:
             xstring = row[0]
             ystring = row[1]
             zstring= row[2]
             X.append(float(xstring))
             Y.append(float(ystring))
             Z.append(float(zstring))
ax.scatter3D(X,Y,Z);

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
plt.show()

def animate(frame):
  ax.view_init(30, frame/4)
  plt.pause(.001)
  return fig

anim = animation.FuncAnimation(fig, animate, frames=200, interval=50)
HTML(anim.to_html5_video())

```
