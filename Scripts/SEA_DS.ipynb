```python
pd_df={"Class":[]};
for i in range(0,10):
  for j in range(0,1000):
    pd_df["Class"].append(dataset[i][j]['Scaler']+dataset[i][j]['Classifier'][0])
pd_unique = set(pd_df['Class'])
print(pd_unique)
```

```python
#HoeffdingTreeClassifier_max_depth
df_param = {};
for i in range(0,1000):
  df_param[i]=[]
print(df_param)
```

```python
#HoeffdingTreeClassifier_max_depth
for i in range(0,1000):
  for j in range(0,10):
    curr = dataset[j][i]['Classifier'][0]
    if(curr=='HoeffdingTreeClassifier'):
      df_param[i].append([dataset[j][i]['Classifier'][1]['max_depth']])
print(df_param)
```

```python
#HoeffdingTreeClassifier_max_depth
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

count_dict=defaultdict(int)
t_output=[]

plt.figure(figsize=(100, 30)) 
xy = np.array([
    (x, y) for x, lst in df_param.items()
    for sublst in lst for y in sublst
])
for x, lst in df_param.items():
  #print(x,lst)
  #to reset the count for each row
  count_dict=defaultdict(int)
  for sublst in lst:
    #print(sublst)
    for y in sublst:
      count_dict[y]+=1
      t_output.append(count_dict[y])
t_output=1000*np.array(t_output)

plt.scatter(*xy.T, s=t_output, color='g')
plt.title("HoeffdingTreeClassifier:max_depth",fontsize=80)
 
# Setting the x and y labels
plt.xlabel("Iteration",fontsize=80)
plt.ylabel("value",fontsize=80)

# Setting the number of ticks
plt.xticks(np.arange(0, len(df_param)+1, 10),fontsize=34, rotation=90)
plt.yticks(fontsize=45)
plt.xlim(xmin=0)
plt.show()
```
