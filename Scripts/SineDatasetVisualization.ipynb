```python
import pandas as pd
```

```python
dataset = pd.read_csv("F:\Laurea Magistrale\Master Thesis\Dataset\Sine\dataset.csv")
```

```python
split_dict_a = {};
split_dict_b = {};
for i in range(1,1001):
    split_dict_a[i]=0;
    split_dict_b[i]=0;
print(split_dict_a);
print(split_dict_b);
```

```python
a=0;
b=1000;
curr=0;
for i in range(1,1001):
    for j in range(a,b):
        curr=curr+dataset.loc[j][0];
    curr=curr/1000;
    split_dict_a[i]=curr;
    a=b+1;
    b=b+1000;
split_dict_a
```

```python
a=0;
b=1000;
curr=0;
for i in range(1,1001):
    for j in range(a,b):
        #print(dataset.loc[j][1]);
        curr=curr+dataset.loc[j][1];
    curr=curr/1000;
    split_dict_b[i]=curr;
    a=b+1;
    b=b+1000;
split_dict_b
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(100, 30)) 
plt.plot(list(split_dict_a.keys()), list(split_dict_a.values()), color='g')
#plt.scatter(list(split_dict_b.keys()), list(split_dict_b.values()), s=800,color='r')
plt.title("Sine Data Visualization",fontsize=80)
 
# Setting the x and y labels
plt.xlabel("Number of Iterations",fontsize=60)
plt.ylabel("Feature value",fontsize=60)
# Setting the number of ticks
plt.xticks(np.arange(1, len(split_dict_b)+1, 10),fontsize=34, rotation=90)
plt.yticks(fontsize=45)
plt.xlim(xmin=0)
plt.show()
```
