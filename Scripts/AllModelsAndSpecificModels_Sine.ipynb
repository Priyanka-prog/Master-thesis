```python
import pandas as pd

pd_df = {"Class": []}

for i in range(0, 10):
    for j in range(0, 1000):
        pd_df["Class"].append(model[i][j]["Scaler"] + str(model[i][j]["Classifier"]))
pd_unique = set(pd_df['Class'])
# Convert dictionary to DataFrame
df = pd.DataFrame(pd_unique)
# Write DataFrame to CSV file
df.to_csv("ALLMODELS.csv", index=False)
```

```python
import pandas as pd

pd_df = {"Class": []}

for i in range(0, 10):
    for j in range(0, 1000):
        curr=model[i][j]["Scaler"] + model[i][j]["Classifier"][0]
        #print(curr)
        if(curr=='MinMaxScalerHoeffdingTreeClassifier'):
           pd_df["Class"].append(model[i][j]["Scaler"] + str(model[i][j]["Classifier"])) 
pd_unique = set(pd_df['Class'])
# Convert dictionary to DataFrame
df = pd.DataFrame(pd_unique)
# Write DataFrame to CSV file
df.to_csv("MinmaxHoeff_Models.csv", index=False)
```

```python
colors = ['#FF5733', '#33FFC1', '#C133FF', '#FF33D1', '#33FF6B', '#FFC133', '#336BFF', '#FF33A8', '#33D1FF', '#FF6B33', '#FF3333', '#33FF33', '#3333FF', '#FFFF33', '#FF33FF', '#33FFFF', '#996633', '#663399']
import re
for p in range(len(pop_model)):
  count_dict = {};
  for i in range(0,1000):
    count_dict[i]=0
  for i in range(0,1000):
    count = 0
    for j in range(0,10):
      curr = model[j][i]['Scaler']+str(model[j][i]['Classifier'])
      pop = str(pop_model.loc[p][0])
      if pop == curr:
        count= count+1
    count_dict[i]=count
  print(pop)
  #print(count_dict)
  import matplotlib.pyplot as plt
  import numpy as np
  pos = np.arange(len(count_dict.keys()))
  plt.figure(figsize=(100, 30))  # width:100, height:30
  plt.bar(range(len(count_dict)), count_dict.values(), align='edge', width=1,color=colors[p])
  pop = re.sub(r'\W+', '_', pop)
  plt.title(pop,fontsize=60)
   
  # Setting the x and y labels
  plt.xlabel("Number of Iterations",fontsize=45)
  plt.ylabel("Number of appearances in each iteration",fontsize=45)
    
  # Setting the number of ticks
  plt.xticks(np.arange(0, len(count_dict)+1, 10),fontsize=34, rotation=90)
  plt.yticks( fontsize=34, rotation=90)
  plt.xlim(xmin=0)
  plt.savefig('plot_{}.png'.format(pop))
  plt.show()
```
