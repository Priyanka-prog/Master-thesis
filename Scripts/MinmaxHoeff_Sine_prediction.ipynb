```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrameXG
prediction_LRR = pd.read_csv("F:\Laurea Magistrale\Master Thesis\Dataset\Sine\Results\minmax_hoeff_dt_predict\prediction.csv", low_memory=False)

# Extract the "real" and "pred" values as separate lists
real = prediction_LRR["real"].tolist()
pred = prediction_LRR["pred"].tolist()

# Create a list of row numbers for the x-axis
x = list(range(1, len(real) + 1))

# Create a scatter plot
plt.scatter(x, real, label="Real", s=10)
plt.scatter(x, pred, label="Pred", s=10)

# Add a legend and axis labels
plt.legend(bbox_to_anchor =(1, 0.5))
plt.title("Decision Tree")
plt.xlabel("Instance")
plt.ylabel("Accuracy_value")

# Show the plot
plt.show()
```
